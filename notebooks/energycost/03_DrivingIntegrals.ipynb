```python
from ebtorch.typing import realnum
from typing import List
import math
from scipy.integrate import quad
```

```python
#
# EXTRACTED COEFFICIENTS
#

Tmax: realnum = 1200

# sqrt(2) * hbar
sqrt2hbar: realnum = (2**0.5) * 1


# NN
nn_a: List[realnum] = [4.07684326171875]
nn_nu: List[realnum] = [4.544354438781738]
nn_phi: List[realnum] = [1.1717623472213745]
nn_b: List[realnum] = [-3.6247568130493164]
nn_mu: List[realnum] = [-3.8860647678375244]
nn_theta: List[realnum] = [-4.871514797210693]

# Star
star_a: List[realnum] = [6.436912536621094, 6.436912536621094]
star_nu: List[realnum] = [4.396866321563721, 0.22187376022338867]
star_phi: List[realnum] = [6.234938621520996, 0.04565233364701271]
star_b: List[realnum] = [-0.12962977588176727, 3.5619561672210693]
star_mu: List[realnum] = [-4.691190242767334, 3.2906980514526367]
star_theta: List[realnum] = [-2.304356575012207, 4.236654758453369]

# FMO
fmo_a: List[realnum] = [4.762247562408447, 4.762247562408447]
fmo_nu: List[realnum] = [0.12661565840244293, 4.125912666320801]
fmo_phi: List[realnum] = [1.5429043769836426, 0.4682064950466156]
fmo_b: List[realnum] = [1.6643606424331665, -0.9458399415016174]
fmo_mu: List[realnum] = [1.1092835664749146, -0.6470829248428345]
fmo_theta: List[realnum] = [-0.5002992153167725, -0.24073514342308044]
```

```python
def integrate_absolute_sum_of_sines(
    amplitudes: List[realnum],
    frequencies: List[realnum],
    phases: List[realnum],
    tt: realnum,
    leading: realnum = 1,
) -> realnum:
    if not (len(amplitudes) == len(frequencies) == len(phases)):
        raise ValueError("All input lists must have the same length.")

    def integrand(t: realnum) -> realnum:
        ssf: realnum = 0.0
        for A, w, p in zip(amplitudes, frequencies, phases):
            ssf += A * math.sin(w * t + p)
        return abs(ssf)

    # noinspection PyTupleAssignmentBalance
    integral, error = quad(integrand, 0, tt, limit=100000)
    return leading * integral
```

```python
# Antenna: NN
print(integrate_absolute_sum_of_sines(nn_a, nn_nu, nn_phi, Tmax, sqrt2hbar))
```

```python
# Antenna: Star
print(integrate_absolute_sum_of_sines(star_a, star_nu, star_phi, Tmax, sqrt2hbar))
```

```python
# Antenna: FMO
print(integrate_absolute_sum_of_sines(fmo_a, fmo_nu, fmo_phi, Tmax, sqrt2hbar))
```

```python
# Sink: NN
print(integrate_absolute_sum_of_sines(nn_b, nn_mu, nn_theta, Tmax))
```

```python
# Sink: Star
print(integrate_absolute_sum_of_sines(star_b, star_mu, star_theta, Tmax))
```

```python
# Sink: FMO
print(integrate_absolute_sum_of_sines(fmo_b, fmo_mu, fmo_theta, Tmax))
```
